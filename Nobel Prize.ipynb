## Proje Soruları

1. Nobel Ödüllerini en çok kazanan ilk on ülkeyi bulunuz.
2. Nobel Ödüllerini kazanan ilk kadınları listeleyiniz.
3. Nobel Ödüllerini kazanan ilk erkekleri listeleyiniz.
4. Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?
5. Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.
6. 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.
7. 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)
8. Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.


```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("nobel.csv")
df.head()
```

```python
# 1- Nobel Ödüllerini en çok kazanan ilk on ülkeyi bulunuz.

df['birth_country'].value_counts().head(10)
```

```python
# 2- Nobel Ödüllerini kazanan ilk kadınları listeleyiniz.

#ilk 10 kadını listeledim ve sadece görmek istediğim 4 sütunu seçtim

df_female= df[df['sex']=='Female'].sort_values(by="year")
df_female_winners = df_female.drop_duplicates(subset="full_name",keep="first")
df_female_winners.head(10)[["year","category","full_name","birth_country"]]
```

```python
# 3- Nobel Ödüllerini kazanan ilk erkekleri listeleyiniz.

#ilk 10 erkeği listeledim ve sadece görmek istediğim 4 sütunu seçtim

df_male= df[df['sex']=='Male'].sort_values(by="year")
df_male_winners = df_male.drop_duplicates(subset="full_name",keep="first")
df_male_winners.head(10)[["year","category","full_name","birth_country"]]
```

```python
# 4- Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip 
# bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?
```

```python
import matplotlib.pyplot as plt

df['birth_country'].value_counts().head(10)
```

```python
df["decade"]=pd.qcut(df["year"], 12 ,labels=[1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010])
```

```python
df.sample(5)
```

```python
df['usa_born_winner'] = df['birth_country'] == 'United States of America'
usa_winners = df.groupby('decade', as_index=False)['usa_born_winner'].mean() 
usa_winners
```

```python
sns.set()
plt.rcParams['figure.figsize'] = [12,7]

ax = sns.lineplot(x='decade', y='usa_born_winner', data=usa_winners)
```

```python
# 5- Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.
#(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

df["birth_date"]=pd.to_datetime(df["birth_date"])
df["age"]=df["year"]-df["birth_date"].dt.year
```

```python
# 6- 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.
```

```python
df_years1 = df[(df.year >=1938) & (df.year<= 1945)]
df_years1.sample(5)
```

```python
df_years2=df_years1.groupby("category").agg({"full_name":"count"})
```

```python
df_years2
```

```python
df_years1.groupby(['category']).count().plot(kind='pie', y='full_name', autopct='%1.0f%%', 
                                             title='Prize Distribution by Categories between 1938-1945', 
                                             legend=False)
```

```python
df_years3=df_years1.groupby("birth_country").agg({"full_name":"count"})
```

```python
df_years3
```

```python
df_years1.groupby(['birth_country']).count().plot(kind='pie', y='full_name', autopct='%1.0f%%',
                                                  title='Winner Countries Between 1938 -1945', 
                                                  legend=False)
```
